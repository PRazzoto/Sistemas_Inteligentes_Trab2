```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
dataset = pd.read_csv('treino_sinais_vitais_com_label.txt')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', splitter="best")
classifier.fit(X_train, y_train)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
y_pred = classifier.predict(X_test)
evaluation_report = classification_report(y_test, y_pred)
print(evaluation_report)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

feature_names = ['pSist','pDiast','qPA','pulso','respiração','gravidade']
class_names = [str(i) for i in np.unique(y)]


plt.figure(figsize=(12,8))
plot_tree(classifier,feature_names=feature_names, class_names=class_names, filled=True, label="root")
plt.savefig("tree_visualization.png", format="png")  # Salva a imagem em formato PNG
plt.show()


```
